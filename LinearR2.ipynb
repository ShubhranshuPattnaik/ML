```python
import pandas as pd
import numpy as np
import sklearn.linear_model
from matplotlib import pyplot as plt
```

```python
reg = sklearn.linear_model.LinearRegression()
```

```python
df = pd.read_csv('/content/drive/MyDrive/MLGoogleColabs/homeprices.csv')
```

```python
df
```

```python
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price)
```

```python
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
135.78767123*3300+180616.43835616432
```

```python
d = pd.read_csv('/content/drive/MyDrive/MLGoogleColabs/areas.csv')
d
```

```python
p = reg.predict(d)
```

```python
d['prediction'] = p
```

```python
d
```

```python
reg.score(df[['area']],df.price)
```

```python
plt.scatter(d.area,d.prediction, color = 'red')
plt.plot(d.area,d.prediction)
```
